#### Exercise 2: Partitioning Around Medoids (PAM)  - _(10 Points)_ 

Load the `USArrests.csv` dataset and the following methods which can be used to compute PAM. Please install the `pyclustering` package.

````python
import pandas as pd 
import numpy as np 
from pyclustering.cluster.kmedoids import kmedoids
from scipy.spatial.distance import pdist, cdist, squareform
import matplotlib.pyplot as plt
from random import randrange
````

````python
df = pd.read_csv('USArrests.csv')
df
````

##### __(a) Standardize the columns of the "USArrests" dataset. Explain why we have to standardize the dataset?__

Standardizing makes variables comparable as each column has different unit and range of values.

````python
mu = abs(df['Murder']-df['Murder'].mean())**2
mu_std = np.sqrt(np.mean(mu))
df['Murder'] = (df['Murder']-df['Murder'].mean())/mu_std

ass = abs(df['Assault']-df['Assault'].mean())**2
ass_std = np.sqrt(np.mean(ass))
df['Assault'] = (df['Assault']-df['Assault'].mean())/ass_std

ur = abs(df['UrbanPop']-df['UrbanPop'].mean())**2
ur_std = np.sqrt(np.mean(ur))
df['UrbanPop'] = (df['UrbanPop']-df['UrbanPop'].mean())/ur_std

ra = abs(df['Rape']-df['Rape'].mean())**2
ra_std = np.sqrt(np.mean(ra))
df['Rape'] = (df['Rape']-df['Rape'].mean())/ra_std
df
````

##### __(b) Compute the PAM (`kmedoids`) using `k = 3`. List the cluster and cluster medoids. Describe and explain the results.__

````python
data = df[['Murder', 'Assault', 'UrbanPop', 'Rape']].values

init_medoids = [randrange(50), randrange(50), randrange(50)]
k_medoids = kmedoids(data, init_medoids)

k_medoids.process()

clusters = k_medoids.get_clusters()
medoids = k_medoids.get_medoids()

print("clusters: ", clusters, "medoids: ", medoids)
````

````python
df1 = pd.read_csv('USArrests.csv')
for c in clusters:
    print(df1.loc[c])
````

It seems to cluster based on crime rate. States in cluster 1 have the highest and states in cluster 3 have the lowest.

##### __(c) Define a Python method to compute the measure for the compactness of a PAM clustering. The parameters of the method is the data and the instance of the kmedoids class (`kmedoids_instance`).__

$$TD=\sum\limits^k_{i = 1} TD(C_i) \text{ and } TD(C) = \sum\limits_{p \in C} dist(p, m_c)$$

The medoid $m_c$ is the representative element of the cluster. 

_(Hint: Use the `pdist` or `cdist` method to compute the pairwise distance between $p$ and $m_c$)_

````python
def compactness(data, kmedoids):
    cols = ['Murder', 'Assault', 'UrbanPop', 'Rape']
    clusters = kmedoids.get_clusters()
    medoids = kmedoids.get_medoids()
    td = 0
    for c in clusters:
        c_idx = clusters.index(c)
        medoid_idx = medoids[c_idx]
        # total distances to medoid_i
        distance = sum(cdist(data.loc[c, cols], data.loc[[medoid_idx], cols]))[0]
        td += distance
    return td
````

##### __(d) Use the newly defined function to compute TD of PAM $k = 3$.__

##### __(e) Compute the TD's for PAM with $k = 1, \dots, 10$ and plot the results. Explain the results of the plot.__

````python
data = df[['Murder', 'Assault', 'UrbanPop', 'Rape']].values
tds = []
for i in range(1, 11):
    init_medoids = []
    for j in range(i):
        init_medoids.append(randrange(len(df)))
    k_medoids = kmedoids(data, init_medoids)
    k_medoids.process()
    tds.append(compactness(df, k_medoids))
tds
````

````python
plt.scatter(list(range(11))[1:], tds)
````

The compactness of PAM clustering decreases while increading k - the number of cluster. This is because the more clusters we choose, the better medoid we can find to assign a data point (medoid that is closer to it). As compactness measures how close are the objects within the same cluster. Lower compactness indicates better clustering.

The sklearn class `GaussianMixture`can be used to compute the EM algorithm. 

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.mixture import GaussianMixture
````

##### __(a) Cluster the `em-data.csv` data. Describe each step and reason your parameter choices.__

````python
em = pd.read_csv('em-data.csv')
em
````

````python
plt.scatter(em['d0'], em['d1'])
````

We can see 3 obvious clusters just by looking at the plot, and the rest data points form the fourth cluster. So we choose the parameter (number of clusters) k = 4.

````python
em_model = GaussianMixture(n_components=4)
em_model.fit(em.values)
y_pred = em_model.predict(em.values)
print(y_pred)
````
